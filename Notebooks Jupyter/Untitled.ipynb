```python
# Libreras externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np


# Librera de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

R1 = 1
R2 = 1 
R3 = 10000
C1 = 0.000001
k = R1/R2

my_tf = TransferFunction( [1,-(k/(C1*R3))], [1,1/(R3*C1)] )


plt.close('all')

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(k) )

pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(k)) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(k))
```

Que utilidad podria tener este tipo de circuitos?
    Estos filtros nos sirven para poder dejar pasar las frecuencias que nosotros queremos, pudiendo asi garantizar que no se mezcle con ruido.
