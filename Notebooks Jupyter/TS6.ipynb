    Se debe diseñar un filtro pasa-altos, que presente máxima planicidad en la banda de paso (frecuencia de corte = 300 Hz) y un cero de transmisión en 100 Hz. El prototipo pasabajos normalizado presenta la siguiente respuesta:

- Determine la expresión de H(s) del filtro pasa-altos normalizado

Analizando los datos que nos arroja el ejercicio sabemos que:
- Frecuencia anulada -> Dos ceros de transmisión en +/- jw
- La transferencia presenta un descenso en la parte de alta frecuencia. Esto quiere decir que la funcion decae a medida que aumentamos la frecuencia. Debido a esto, el numero de polos sera mayor que el de ceros. 
- Siendo que tenemos dos ceros, tendremos tres polos.

Viendo el diagrama de bode del ejercicio, los polos se encuentran en +/- 3j

Por caso, la transferencia nos queda:

\begin{align*}
 T_L (S) = \frac{{S^2+9}}{{(S+1)(S^2+2\cos(\frac{\pi}{3})+1)}} 
\end{align*}

Debo multiplicar por un factor k = 1/9 para tener ganancia unitaria. Con esto, la transferencia es:

\begin{align*}
 T_L (S) = \frac{{\frac{1}{9}S^2+1}}{{S^3+2S^2+2S+1}} 
\end{align*}

Chequeamos mediante python que la transferencia se corresponda al grafico propuesto en el enunciado:

```python
import numpy as np
from pytc2.sistemas_lineales import TransferFunction, pzmap, bodePlot

num = [1/9,0,1]
den = [1,2,2,1]

T_L = TransferFunction(num,den)

bodePlot(T_L)
pzmap(T_L)
```

- Sintetice el circuito del filtro pedido. Se utilizará para la estructura de segundo orden el siguiente circuito:

```python
from sympy import symbols, Eq, solve

# Variables
VA, VB, Vi, Vo, s, C, G, G1, G2 = symbols('VA VB Vi Vo s C G G1 G2')

# Ecuaciones
eq1 = Eq(VA*(s*C+G1+G2) - Vi*s*C - VB*G2, 0)
eq2 = Eq(VA*(G+G) - VB*G - Vo*G, 0)
eq3 = Eq(VA*(s*C+G) - Vo*s*C - Vi*G, 0)

# Resolver el sistema de ecuaciones
sol = solve((eq1, eq2, eq3), (VA, VB, Vi, Vo))

# Mostrar la solución
print("Solución:")
print(f"Vi = {sol[Vi]}")
```

Por lo tanto, nos queda:

\begin{align*}
 \frac{Vo}{Vi} = \frac{{s^2+\frac{-GG_1+GG_2}{C^2}}}{{s^2+s\frac{G_1}{C}+\frac{G_2G}{C^2}}} 
\end{align*}

Analizando el resultado:
- Nos queda una transferencia de segundo orden con dos polos complejos conjugados y dos ceros
- Wo de los ceros nos queda $\sqrt{\frac{-GG_1+GG_2}{C^2}}$
- Wo de los polos nos queda $\sqrt{\frac{GG_2}{C^2}}$

Haciendo la relacion entre los cuadrados de ambos Wo obtenemos:
- $ G1 = G2(1-a) $

- Donde $ a = (\frac{W_oz}{W_op})^2 $ , por lo que $ a > 0 $ , entonces:
$ G1 < G2 $
